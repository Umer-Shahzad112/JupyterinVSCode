```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dots = sns.load_dataset('dots')
dots.head()
```

```python
palette  = sns.set_palette('rocket_r')
```

```python
sns.relplot(data= dots, x = 'time', y = 'firing_rate', size = 'choice', col = 'align', kind = 'line',palette = palette, height = 5, aspect = .8, facet_kws=dict(sharex =False))
```

```python
sns.relplot(data= dots, x = 'time', y = 'firing_rate', hue = 'coherence',size = 'choice',size_order= ['T1','T2'],
col = 'align', kind = 'line',palette = palette, height = 5, aspect = .75, facet_kws=dict(sharex =False))
```

```python
tip = sns.load_dataset('tips')
sns.boxplot(x = 'day',y= 'total_bill', hue = 'smoker', data = tip, palette = ['m','g'])
sns.despine(offset=10 , trim= True)
```

```python
sns.violinplot(data =tip, x = 'day',y= 'total_bill', hue = 'smoker', split =True, inner = 'quart',linewidth=2,
palette= {'Yes':'b', 'No':'.85'})
sns.despine(left= True)
```

```python
sns.set_style('whitegrid')
diamonds = sns.load_dataset('diamonds')
diamonds.value_counts('clarity')
f, ax = plt.subplots(figsize = (60,60))
sns.despine(f , left =True, bottom= True)
sns.scatterplot(x = 'carat', y = 'price', hue = 'clarity' , size = 'depth', data = diamonds, palette = 'ch: r=-.2, d = .3_r',
sizes = (1 ,100), linewidth =0, ax = ax)
plt.legend()
```

```python
planets = sns.load_dataset('planets')
f , ax = plt.subplots(figsize=(15,7))
ax.set_xscale('log')
sns.boxplot(x = 'distance', y= 'method',data = planets, whis =[0,100], width = .6, palette = 'vlag')

sns.stripplot(x = 'distance', y= 'method',data = planets,size = 4, color = '.3', linewidth=0)
ax.xaxis.grid(True)
```
