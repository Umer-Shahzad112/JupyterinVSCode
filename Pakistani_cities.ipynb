```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.set_theme(style="ticks", color_codes=True)
cord = pd.read_csv('cord.csv')
```

```python
plt.figure(figsize=(50,50))
p1 = sns.scatterplot(x= 'Longitude', y= 'Latitude', data= cord)

for line in range(0,cord.shape[0]):
    p1.text(cord.Longitude[line]+.01, cord.Latitude[line], 
    cord.City[line], horizontalalignment='left', 
    size='large', color='black', weight='semibold')


plt.show()
```
