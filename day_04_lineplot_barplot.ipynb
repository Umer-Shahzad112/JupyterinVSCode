```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
sns.lineplot(x = 'age', y='fare', data=titanic)
plt.title('Age vs Fare')
```

```python
sns.lineplot(x = 'age', y='fare', data=titanic)
plt.title('Age vs Fare')

plt.ylabel('Fare')
plt.xlabel('Age')

plt.xlim(right = 66)
plt.ylim(3,top = 150)
```

```python
sns.set_style('darkgrid')
sns.lineplot(x = 'age', y='fare', data=titanic)
plt.title('Age vs Fare')
```

```python
plt.figure(figsize=(15,4))
sns.lineplot(x = 'age', y='fare', data=titanic)
plt.title('Age vs Fare')
```

```python
plt.figure(figsize=(15,4))
sns.lineplot(x = 'age', y='fare', data=titanic, color= 'red', palette='dark')
plt.title('Age vs Fare')
```
